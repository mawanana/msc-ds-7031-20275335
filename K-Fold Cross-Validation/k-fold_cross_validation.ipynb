```python
import pandas as pd
from sklearn.model_selection import KFold

# Load your dataset
df = pd.read_csv('Master_Dataset.csv')

# Define the number of folds (K)
n_splits = 5

# Initialize K-Fold Cross-Validation
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize a counter to create unique file names
file_counter = 1

# Loop through the K-Folds
for train_index, test_index in kf.split(df):
    train_data = df.iloc[train_index]
    test_data = df.iloc[test_index]

    # Save the training and test datasets as CSV files
    train_data.to_csv(f'train_data_{file_counter}.csv', index=False)
    test_data.to_csv(f'test_data_{file_counter}.csv', index=False)

    # Print the first few rows of the training and test datasets for this fold
    print(f"Fold {file_counter}:")
    print("Training Dataset:")
    print(train_data.head())
    print("Test Dataset:")
    print(test_data.head())

    file_counter += 1
print("Done")
```

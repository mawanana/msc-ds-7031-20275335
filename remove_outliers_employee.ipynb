```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("data/Cleaned_Employee.csv")

# Convert the "DateofJoin" column to datetime
df["DateofJoin"] = pd.to_datetime(df["DateofJoin"])

# Create a subplot for density and box plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Plot density plots for numeric columns (e.g., DateofJoin)
sns.kdeplot(data=df, x="DateofJoin", ax=axes[0])
axes[0].set_title("Density Plot for DateofJoin")

# Plot box plots to identify outliers
sns.boxplot(data=df, x="DateofJoin", ax=axes[1])
axes[1].set_title("Box Plot for DateofJoin")

# Display the plots
plt.show()

# Define a threshold for outlier detection
dateofjoin_threshold = df["DateofJoin"].quantile(0.99)

# Remove outliers based on the threshold
df = df[df["DateofJoin"] <= dateofjoin_threshold]

# Save the cleaned DataFrame back to a CSV file
df.to_csv("data/Cleaned_Employee_NoOutliers.csv", index=False)
```

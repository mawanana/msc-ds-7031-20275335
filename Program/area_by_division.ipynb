```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("data/Fields.csv")

# Group the data by 'Division' and calculate the sum of 'Area (Ha)' for each division
division_area = df.groupby("Division")["Area (Ha)"].sum()

# Calculate the percentage of each division's area
total_area = division_area.sum()
division_percentage = (division_area / total_area) * 100

# Create a bar chart with percentage values
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
bars = plt.bar(division_area.index, division_area)
plt.xlabel("Division")
plt.ylabel("Area (Ha)")
plt.title("Area (Ha) by Division")

# Add percentage labels on top of each bar
for bar, percentage in zip(bars, division_percentage):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f"{percentage:.2f}%", ha='center', va='bottom')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.show()
```

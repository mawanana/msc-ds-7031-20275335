```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("data/Fields.csv")

# Group the data by "Division" and calculate the sum of "NumberOfTree"
division_tree_sum = df.groupby("Division")["NumberOfTree"].sum()

# Calculate the total number of trees across all divisions
total_trees = division_tree_sum.sum()

# Plot the bar chart
plt.figure(figsize=(10, 6))
ax = division_tree_sum.plot(kind="bar")
plt.xlabel("Division")
plt.ylabel("NumberOfTree")
plt.title("Number of Trees by Division")
plt.xticks(rotation=45)

# Add percentage labels to each bar
for i, value in enumerate(division_tree_sum):
    percentage = (value / total_trees) * 100
    ax.text(i, value, f"{percentage:.2f}%", ha='center', va='bottom')

plt.show()
```

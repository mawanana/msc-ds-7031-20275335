```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("data/Merged_Fields_NumberOfEmp.csv")

# Group the data by 'Division' and calculate the mean values
grouped_df = df.groupby('Division')[['Area (Ha)', 'NumberOfTree', 'NumberOfEmp']].mean()

# Calculate the percentage values for each column
grouped_df['Area (Ha)'] = (grouped_df['Area (Ha)'] / grouped_df['Area (Ha)'].sum()) * 100
grouped_df['NumberOfTree'] = (grouped_df['NumberOfTree'] / grouped_df['NumberOfTree'].sum()) * 100
grouped_df['NumberOfEmp'] = (grouped_df['NumberOfEmp'] / grouped_df['NumberOfEmp'].sum()) * 100

# Plot the multi-bar chart
grouped_df.plot(kind='bar', figsize=(10, 6))
plt.title("Percentage of Area (Ha), NumberOfTree, and NumberOfEmp by Division")
plt.xlabel("Division")
plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.legend(["Area (Ha)", "NumberOfTree", "NumberOfEmp"])
plt.show()
```

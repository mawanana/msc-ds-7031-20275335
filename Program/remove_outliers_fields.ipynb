```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Load the dataset
data = pd.read_csv("data/Cleaned_Fields.csv")

# Create density plot
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.kdeplot(data['Area (Ha)'], shade=True)
plt.title("Density Plot for 'Area (Ha)'")
plt.show()

# Create box plot
plt.figure(figsize=(8, 4))
sns.boxplot(data=data, x='Area (Ha)')
plt.title("Box Plot for 'Area (Ha)'")
plt.show()

# Calculate Z-scores
z_scores = stats.zscore(data['Area (Ha)'])
abs_z_scores = abs(z_scores)
threshold = 3  

# Remove outliers based on Z-scores
data_no_outliers = data[(abs_z_scores < threshold)]

# Save the cleaned data to a new CSV file
data_no_outliers.to_csv("data/Cleaned_Fields_NoOutliers.csv", index=False)

print("Outliers removed and cleaned data saved as 'Cleaned_Fields_NoOutliers.csv'.")
```

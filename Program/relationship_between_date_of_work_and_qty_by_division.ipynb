```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

# Read the CSV file into a DataFrame
df = pd.read_csv("data/WorkDetails_modified.csv")

# Convert 'DateofWork' to datetime
df['DateofWork'] = pd.to_datetime(df['DateofWork'])

# Group the data by Division
grouped = df.groupby("Division")

# Create a scatter plot for each division
fig, ax = plt.subplots(figsize=(12, 6))

for name, group in grouped:
    ax.scatter(group["DateofWork"], group["Qty"], label=name)

ax.set_xlabel("Date of Work")
ax.set_ylabel("Qty")
ax.set_title("Relationship between Date of Work and Qty by Division")

# Rotate x-axis labels for better visibility
date_format = DateFormatter("%m/%d/%Y")
ax.xaxis.set_major_formatter(date_format)
plt.xticks(rotation=45)

ax.legend()
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("data/WorkDetails_modified.csv")

# Calculate the count of null values in each column
null_counts = df.isnull().sum()

# Create a bar chart with black bars
null_counts.plot(kind='bar', figsize=(10, 6), color='black')
plt.title('Null Value Counts in Each Column')
plt.xlabel('Columns')
plt.ylabel('Null Value Count')
plt.xticks(rotation=45)
plt.show()
```

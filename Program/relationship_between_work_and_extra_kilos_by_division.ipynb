```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("data/WorkDetails_modified.csv")

# Group the data by Division
grouped = df.groupby("Division")

# Create a scatter plot for "Work" and "ExtraKilos" for each division
plt.figure(figsize=(10, 6))

for name, group in grouped:
    plt.scatter(group["Work"], group["ExtraKilos"], label=name)

plt.xlabel("Work")
plt.ylabel("Extra Kilos")
plt.title("Relationship between Work and Extra Kilos by Division")
plt.legend()

plt.show()
```

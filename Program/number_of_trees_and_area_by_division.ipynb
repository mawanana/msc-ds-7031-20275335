```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("data/Fields.csv")

# Group the data by "Division" and calculate the sum of "NumberOfTree" and "Area (Ha)"
division_tree_area_sum = df.groupby("Division")[["NumberOfTree", "Area (Ha)"]].sum()

# Calculate the total number of trees and total area across all divisions
total_trees = division_tree_area_sum["NumberOfTree"].sum()
total_area = division_tree_area_sum["Area (Ha)"].sum()

# Plot the multi-bar chart
plt.figure(figsize=(10, 6))
ax = division_tree_area_sum.plot(kind="bar", secondary_y=["Area (Ha)"])

# Set y-axis labels and title
plt.xlabel("Division")
plt.title("Number of Trees and Area by Division")

# Set primary y-axis (Number of Trees) labels and percentages
ax.set_ylabel("Number of Trees")
for i, value in enumerate(division_tree_area_sum["NumberOfTree"]):
    percentage = (value / total_trees) * 100
    ax.text(i, value, f"{percentage:.2f}%", ha='center', va='bottom')

# Set secondary y-axis (Area) labels and percentages
plt.ylabel("Area (Ha)")
for i, value in enumerate(division_tree_area_sum["Area (Ha)"]):
    percentage = (value / total_area) * 100
    plt.text(i, value, f"{percentage:.2f}%", ha='center', va='top')

# Display the legend
plt.legend(loc='upper left')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

plt.show()
```

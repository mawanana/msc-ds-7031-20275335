```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("data/Merged_Employee_WorkDetails.csv")

# Group the data by "DateofWork" and calculate the mean for each group
grouped = df.groupby("DateofWork")["Qty"].mean().reset_index()

# Create a line chart with the mean "Qty" values for each "DateofWork"
plt.figure(figsize=(12, 6))
plt.plot(grouped["DateofWork"], grouped["Qty"])
plt.xlabel("DateofWork")
plt.ylabel("Mean Qty")
plt.title("Mean Qty for Each DateofWork")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

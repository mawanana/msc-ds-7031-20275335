```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('data/Employee.csv')

# Group the data by Division and Gender, and count the occurrences
grouped = data.groupby(['Division', 'Gender']).size().unstack().fillna(0)

# Create a multi-bar graph with separate bars for male and female
ax = grouped.plot(kind='bar', figsize=(10, 6), width=0.4, position=0.5)
plt.title("Employee Gender Distribution by Division")
plt.xlabel("Division")
plt.ylabel("Count")
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

# Show the legend
plt.legend(title="Gender", labels=["Female", "Male"])

# Show the plot
plt.show()
```

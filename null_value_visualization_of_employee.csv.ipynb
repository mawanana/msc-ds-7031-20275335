```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file into a DataFrame
data = pd.read_csv('data/Employee.csv')

# Create a custom colormap with the specified colors
custom_cmap = sns.color_palette(["black", "white"])

# Generate a heatmap to visualize null values using the custom colormap
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap=custom_cmap)
plt.title("Null Value Visualization of Employee.csv")
plt.show()
```

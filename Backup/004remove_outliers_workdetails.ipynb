```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("data/Cleaned_WorkDetails.csv")

# Create density plots
sns.kdeplot(data['Qty'], label="Qty")
sns.kdeplot(data['ExtraKilos'], label="ExtraKilos")
plt.legend()
plt.show()

# Create box plots
sns.boxplot(data=data[['Qty', 'ExtraKilos']])
plt.show()
```

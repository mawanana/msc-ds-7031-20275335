```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Load the dataset
data = pd.read_csv("data/Cleaned_WorkDetails.csv")

# Create density plot
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.kdeplot(data['Workdayfraction'], shade=True)
plt.title('Density Plot for Workdayfraction')
plt.show()

# Create box plot
plt.figure(figsize=(8, 4))
sns.boxplot(data=data, x='Workdayfraction')
plt.title('Box Plot for Workdayfraction')
plt.show()

z_scores = stats.zscore(data['Workdayfraction'])
abs_z_scores = abs(z_scores)
threshold = 3  # You can adjust the threshold as needed
outliers = (abs_z_scores > threshold)

# Remove outliers
data_cleaned = data[~outliers]

# Save the cleaned dataset to 'Cleaned_WorkDetails_NoOutliers.csv'
data_cleaned.to_csv('data/Cleaned_WorkDetails_NoOutliers.csv', index=False)

print("Done")
```

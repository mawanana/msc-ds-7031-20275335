```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
data = pd.read_csv("data/Cleaned_WorkDetails.csv")

# Visualize the data (Optional)
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=data[['Qty', 'ExtraKilos']])
plt.show()

sns.kdeplot(data['Qty'], fill=True)  
sns.kdeplot(data['ExtraKilos'], fill=True)
plt.show()

# sns.kdeplot(data['Qty'], shade=True)
# sns.kdeplot(data['ExtraKilos'], shade=True)
# plt.show()

# Calculate Z-scores
z_scores = stats.zscore(data[['Qty', 'ExtraKilos']])
abs_z_scores = np.abs(z_scores)
threshold = 3  # Adjust the threshold as needed

# Identify good records (not outliers)
good_records_mask = (abs_z_scores < threshold).all(axis=1)
good_records = data[good_records_mask]

# Create a DataFrame of outlier records
outliers = data[~good_records_mask]

# Save the good records to a new CSV file
good_records.to_csv("data/Cleaned_WorkDetails_GoodRecords.csv", index=False)

# Save the outlier records to a new CSV file
outliers.to_csv("data/Z-scores_outliers.csv", index=False)
```

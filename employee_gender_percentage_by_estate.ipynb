```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('data/Employee.csv')

# Group the data by Estate and Gender, and count the occurrences
grouped = data.groupby(['Estate', 'Gender']).size().unstack().fillna(0)

# Calculate the total number of employees (male + female) for each estate
total = grouped['F'] + grouped['M']

# Calculate the percentage of females and males in each estate
grouped['Female %'] = (grouped['F'] / total) * 100
grouped['Male %'] = (grouped['M'] / total) * 100

# Create a bar graph to show the percentage of females and males for each estate
ax = grouped[['Female %', 'Male %']].plot(kind='bar', figsize=(10, 6))
plt.title("Employee Gender Percentage by Estate")
plt.xlabel("Estate")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

# Annotate the bars with percentage values
for i, v in enumerate(grouped['Female %']):
    ax.text(i, v + 2, f'{v:.2f}%', ha='center', va='bottom', color='black')
for i, v in enumerate(grouped['Male %']):
    ax.text(i, v + 2, f'{v:.2f}%', ha='center', va='bottom', color='black')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('data/Employee.csv')

# Remove rows with NULL join date
data = data.dropna(subset=['DateofJoin'])

# Convert the 'DateofJoin' column to datetime
data['DateofJoin'] = pd.to_datetime(data['DateofJoin'], errors='coerce')

# Create a line graph for each division
plt.figure(figsize=(10, 6))

# Get unique divisions
divisions = data['Division'].unique()

# Define colors for each division
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

for i, division in enumerate(divisions):
    division_data = data[data['Division'] == division]
    division_data = division_data.groupby('DateofJoin').size()
    division_data.plot(label=division, color=colors[i])

plt.title('Number of Employees by Division')
plt.xlabel('Join Date')
plt.ylabel('Number of Employees')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```
